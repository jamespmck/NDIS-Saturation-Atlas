# NDIS Market Saturation Analysis

This notebook is the reproducible analytical companion to the Streamlit atlas. It uses the curated published CSV files and recreates the core benchmark logic in a compact, reviewable form.

The aim is not to produce a definitive ranking of need. The aim is to identify service areas whose plan coverage, utilisation or service-category payment mix differs enough from a selected benchmark to justify closer local interpretation.

## 1. Setup

The notebook deliberately relies on the same curated files as the app rather than rebuilding the raw-data pipeline. This keeps the audit trail focused on the analytical layer.

```python
from pathlib import Path

import pandas as pd

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == 'notebooks':
    PROJECT_ROOT = PROJECT_ROOT.parent

PUBLISHED = PROJECT_ROOT / 'data' / 'published'
MASTER_PATH = PUBLISHED / 'master_ndis_service_area_quarter_all_available_scoped.csv'
SERVICE_TYPE_PATH = PUBLISHED / 'master_ndis_service_area_quarter_service_type_custom.csv'

PLAN_COL = 'service_area_funded_plans_per_1000_population_2025_erp'
UTIL_COL = 'service_area_mean_plan_utilisation'
PLAN_COUNT_COL = 'funded_plans_count'
POP_COL = 'population_2025_erp'

PLAN_GAP_COL = 'funded_plans_per_1000_gap_from_national'
UTIL_GAP_COL = 'mean_plan_utilisation_gap_from_national'
```

## 2. Load Curated Data

```python
master = pd.read_csv(MASTER_PATH, low_memory=False)
service_mix = pd.read_csv(SERVICE_TYPE_PATH, low_memory=False)

print(f'Master rows: {len(master):,}')
print(f'Service-category rows: {len(service_mix):,}')
print(f'Quarters: {master['quarter'].nunique():,}')
print(f'Service areas: {master['ndis_service_area'].nunique():,}')

master.head()
```

## 3. Validate Required Columns

A lightweight schema check is enough for notebook exploration. The app and tests provide the more durable regression checks.

```python
required_master = {
    'quarter',
    'map_key',
    'ndis_service_area',
    'remoteness_category',
    POP_COL,
    PLAN_COUNT_COL,
    PLAN_COL,
    UTIL_COL,
}

required_service_mix = {
    'quarter',
    'ndis_service_area',
    'service_type_group',
    'service_type_payment_amount',
    'service_type_payment_share_of_area_total',
}

missing_master = sorted(required_master - set(master.columns))
missing_service_mix = sorted(required_service_mix - set(service_mix.columns))

assert not missing_master, missing_master
assert not missing_service_mix, missing_service_mix

master['quarter'] = master['quarter'].astype(str)
service_mix['quarter'] = service_mix['quarter'].astype(str)

for column in [POP_COL, PLAN_COUNT_COL, PLAN_COL, UTIL_COL]:
    master[column] = pd.to_numeric(master[column], errors='coerce')

for column in ['service_type_payment_amount', 'service_type_payment_share_of_area_total']:
    service_mix[column] = pd.to_numeric(service_mix[column], errors='coerce')

print('Schema check passed')
```

## 4. Recreate National Benchmarks

The plan benchmark is population-weighted. The utilisation benchmark is weighted by funded-plan count because it is a plan-level utilisation context measure.

```python
def weighted_mean(values: pd.Series, weights: pd.Series) -> float:
    values = pd.to_numeric(values, errors='coerce')
    weights = pd.to_numeric(weights, errors='coerce')
    mask = values.notna() & weights.notna() & (weights > 0)
    if not mask.any():
        return float(values.mean())
    return float((values[mask] * weights[mask]).sum() / weights[mask].sum())

benchmarks = (
    master.groupby('quarter', as_index=False)
    .apply(lambda group: pd.Series({
        'national_plans_per_1000': group[PLAN_COUNT_COL].sum() / group[POP_COL].sum() * 1000,
        'national_mean_utilisation': weighted_mean(group[UTIL_COL], group[PLAN_COUNT_COL]),
    }), include_groups=False)
)

analysis = master.merge(benchmarks, on='quarter', how='left')
analysis['plan_gap_from_national'] = analysis['national_plans_per_1000'] - analysis[PLAN_COL]
analysis['utilisation_gap_from_national'] = analysis['national_mean_utilisation'] - analysis[UTIL_COL]

benchmarks.tail()
```

## 5. Latest-Quarter Service-Area Signals

```python
def quarter_key(value: str) -> tuple[int, int]:
    year, quarter = value.split('Q')
    return int(year), int(quarter)

latest_quarter = sorted(analysis['quarter'].dropna().unique(), key=quarter_key)[-1]
latest = analysis.loc[analysis['quarter'].eq(latest_quarter)].copy()

review_columns = [
    'ndis_service_area',
    'service_area_state_label',
    'remoteness_category',
    PLAN_COL,
    'national_plans_per_1000',
    'plan_gap_from_national',
    UTIL_COL,
    'national_mean_utilisation',
    'utilisation_gap_from_national',
]

latest.sort_values('plan_gap_from_national', ascending=False)[review_columns].head(15)
```

## 6. Remoteness Summary

```python
remoteness_summary = (
    latest.groupby('remoteness_category', as_index=False)
    .agg(
        service_areas=('ndis_service_area', 'nunique'),
        population=(POP_COL, 'sum'),
        funded_plans=(PLAN_COUNT_COL, 'sum'),
        mean_plans_per_1000=(PLAN_COL, 'mean'),
        median_plan_gap=('plan_gap_from_national', 'median'),
        mean_utilisation=(UTIL_COL, 'mean'),
        median_utilisation_gap=('utilisation_gap_from_national', 'median'),
    )
    .sort_values('median_plan_gap', ascending=False)
)

remoteness_summary
```

## 7. Service-Category Payment Mix

This section uses payment share as a composition proxy. It does not estimate unique participants by service category.

```python
latest_mix = service_mix.loc[service_mix['quarter'].eq(latest_quarter)].copy()

national_mix = (
    latest_mix.groupby('service_type_group', as_index=False)
    .agg(payment_amount=('service_type_payment_amount', 'sum'))
)
national_mix['national_payment_share'] = national_mix['payment_amount'] / national_mix['payment_amount'].sum()

area_mix = (
    latest_mix.groupby(['ndis_service_area', 'service_area_state_label', 'service_type_group'], as_index=False)
    .agg(
        payment_amount=('service_type_payment_amount', 'sum'),
        payment_share=('service_type_payment_share_of_area_total', 'sum'),
    )
    .merge(national_mix[['service_type_group', 'national_payment_share']], on='service_type_group', how='left')
)
area_mix['payment_share_gap_pp'] = (area_mix['payment_share'] - area_mix['national_payment_share']) * 100

area_mix.sort_values('payment_share_gap_pp', ascending=False).head(15)
```

## 8. Export Review Tables

The export keeps the notebook useful for peer review without requiring the Streamlit app.

```python
output_dir = PROJECT_ROOT / 'outputs' / 'tables'
output_dir.mkdir(parents=True, exist_ok=True)

latest[review_columns].to_csv(output_dir / f'latest_quarter_service_area_review_{latest_quarter}.csv', index=False)
remoteness_summary.to_csv(output_dir / f'remoteness_summary_{latest_quarter}.csv', index=False)
area_mix.to_csv(output_dir / f'service_category_payment_mix_gap_{latest_quarter}.csv', index=False)

print(f'Wrote review tables to {output_dir}')
```
